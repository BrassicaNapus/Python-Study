```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```
df = pd.read_csv('train.csv')
df
```

```
df.info()
```

```
df.dropna(subset=['Age']) # NaN 제거
```

```
df[df['Age'] <= 10]
```

```
df[df['Age'] <= 10][df[df['Age'] <= 10]['Survived'] == 1]
```

```
len(df[df['Age'] <= 10][df[df['Age'] <= 10]['Survived'] == 1]) / len(df[df['Age'] <= 10])
```

```
df[(df['Age'] >= 20) & (df['Age'] <= 30)]
```

```
df[(df['Age'] >= 20) & (df['Age'] <= 30)][df[(df['Age'] >= 20) & (df['Age'] <= 30)]['Survived'] == 1]
```

```
len(df[(df['Age'] >= 20) & (df['Age'] <= 30)][df[(df['Age'] >= 20) & (df['Age'] <= 30)]['Survived'] == 1]) / len(df[(df['Age'] >= 20) & (df['Age'] <= 30)])
```

```
len(df[(df['Age'] >= 20) & (df['Age'] <= 30) & (df['Survived'] == 1)])/len(df[(df['Age'] >= 20) & (df['Age'] <= 30)])
```

```
ages = [[0,9],[10,19],[20,29],[30,39],[40,49],[50,59]]
arr = []
for age in ages:
  a = len(df[(df['Age'] >= age[0]) & (df['Age'] <= age[1]) & (df['Survived'] == 1)])/len(df[(df['Age'] >= age[0]) & (df['Age'] <= age[1])])
  arr.append(a)
print(arr)
```

```
plt.bar(np.arange(len(ages)), arr)
plt.xticks(np.arange(len(ages)),[str(i[0])+'~'+str(i[1]) for i in ages])
plt.show()
```
