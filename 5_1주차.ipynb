```
import numpy as np
import matplotlib.pyplot as plt
```

```
plt.plot([0,np.pi/2,np.pi,np.pi/2*3,np.pi])
plt.show()
```

```
a = np.linspace(0, np.pi*2, 1000)
plt.plot(a, np.sin(a))
plt.show()
```

```
a = np.linspace(-np.pi/2, np.pi/2, 1000)
plt.ylim([-10, 10])
plt.plot(a, np.tan(a))
plt.show()
```

```
a = np.linspace(0, np.pi*2, 1000)
plt.plot(a, np.cos(a))
plt.show()
```

```
a = np.linspace(0, np.pi*2, 1000)
plt.plot(a, np.exp(a))
plt.show()
```

```
a = np.linspace(0, np.pi*2, 1000)
plt.plot(a, np.log(a))
plt.show()
```

```
a = np.linspace(-8, 8, 1000)
plt.plot(a, 1 / (1 + np.exp(-a)))
plt.show()
```
